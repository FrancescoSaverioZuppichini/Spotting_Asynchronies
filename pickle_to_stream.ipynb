# Spacchetta i file .pickle e ne salva gli input stream e target stream nelle directories /DATA_STREAM/mode_[modalit√†]

```python
import numpy as np

from os import listdir
from os.path import isfile, join
```

### Ventilation modes:
1. Continuous mandatory ventilation
2. Intermittent mandatory ventilation
3. Synchronized intermittent mandatory ventilation
4. Continuous spontaneous ventilation
5. Continuous spontaneous ventilation with NAVA

```python
def which_mode(mode_onehot):
    
    if mode_onehot == (1, 0, 0, 0, 0):
        return 0
    
    elif mode_onehot == (0, 1, 0, 0, 0):
        return 1
    
    elif mode_onehot == (0, 0, 1, 0, 0):
        return 2
    
    elif mode_onehot == (0, 0, 0, 1, 0):
        return 3
    
    elif mode_onehot == (0, 0, 0, 1, 1):
        return 4
    else:
        print("ERRORE: mode non prevista: ", mode_onehot)
```

```python
directory = "PICKLED_DATA/"
onlyfiles = [f for f in listdir(directory) if isfile(join(directory, f))]

mode_counter = [0]*5 # conteggio degli streams per ogni mode
modes = np.array([0, 1, 2, 3, 4])

"""Per ogni elemento nella cartella dei file .pickle, viene letta la mode, estratto lo stream di input 
    e lo stream di target, e questi due vengono salvati nella relativa cartella mode"""
for pickle_file_name in onlyfiles:
    
    pickled_data = open(directory+pickle_file_name, "rb")
    raw_data = np.load(pickled_data)

    print(raw_data[0][0]) # Printa la mode
    mode = which_mode(raw_data[0][0]) # Lettura mode
    
    data = raw_data[0][1] # Input + target stream

    X = np.array([data[i][0] for i in range(len(data))]) #Estrazione input stream
    y = np.array([data[i][1] for i in range(len(data))]) #Estrazione target stream
    
    saving_folder = "DATA_STREAMS/mode_"+str(mode)+"/"
    input_stream_name = "input_stream_"+str(mode_counter[mode])+".npy"
    target_stream_name = "target_stream_"+str(mode_counter[mode])+".npy"

    np.save(saving_folder + input_stream_name, X)
    np.save(saving_folder + target_stream_name, y)
    
    mode_counter[mode] += 1 # aumenta di uno il conteggio per la modalita' appena registrata

    print(pickle_file_name + " saved as .npy in " + saving_folder)
    print("Ventilation mode: ", mode, "\n")
    
```

```python
print("Number of (input_stream, target_stream) for each mode: ", mode_counter)
```
